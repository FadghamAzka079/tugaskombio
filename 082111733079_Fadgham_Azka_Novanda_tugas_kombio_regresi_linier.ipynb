```python
#contoh soal
import numpy as np
import pandas as pd

# Menginput data x dan y
x = np.array([0.0, 1.0, 2.0, 2.5, 3.0])
y = np.array([2.9, 3.7, 4.1, 4.4, 5.0])

# Menginisiasi variabel
n = len(x)
sumx = 0
sumy = 0
mean_x = np.mean(x)
mean_y = np.mean(y)
xy = np.zeros(n, float)
x_squared = np.zeros(n, float)
si = np.zeros(n, float)

# Looping untuk menghitung data
for a in range(n):
    sumx = sumx + x[a]
    x_squared[a] = x[a] ** 2
    sumy = sumy + y[a]
    xy[a] = x[a] * y[a]

# Menghitung koefisien regresi (slope dan intercept)
b = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
a = mean_y - b * mean_x

# Menghitung s_i (error kuadrat)
for i in range(n):
    si[i] = (y[i] - (a + b * x[i])) ** 2

# Membuat matriks C dan D untuk menghitung koefisien regresi
C = np.array([[n, sumx], [sumx, np.sum(x_squared)]])
D = np.array([[sumy], [np.sum(xy)]])

# Menghitung nilai prediksi dari koefisien regresi sebelumnya
ybaru = np.zeros(n, float)
for i in range(n):
    ybaru[i] = a + b * x[i]

# Menghitung Nilai S
S = np.sum(si)

# Mencetak persamaan regresi
print('y =', b, 'x +', a)

# Buat DataFrame
data = {'No.': range(1, len(x) + 1), 'x': x, 'y': y, 'xy': xy, 'x^2': x_squared, 's_i': si}
df = pd.DataFrame(data)

# Tampilkan seluruh nilai hasil perhitungan
print(df)
print('Nilai a =', a)
print('Nilai b =', b)
print('S =', S)
print('Jumlah X =', sumx)
print('Jumlah XY =', np.sum(xy))
```

```python
import numpy as np
import pandas as pd

# Menginput data t dan y
t = np.array([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5])
y = np.array([1.000, 0.994, 0.990, 0.985, 0.979, 0.977, 0.972, 0.969, 0.967, 0.960, 0.956, 0.952])

# Menginisiasi variabel
n = len(t)
sumt = 0
sumlny = 0
si = np.zeros(n, float)
mean_t = np.mean(t)
mean_ln_y = np.mean(np.log(y))
ln_y = np.log(y)
t_times_ln_y = t * ln_y
t_squared = t ** 2
ln_y_squared = ln_y ** 2

# Looping untuk menghitung data
for a in range(n):
    sumt = sumt + t[a]
    sumlny = sumlny + ln_y[a]

# Menghitung koefisien regresi (slope dan intercept)
b = (n * np.sum(t_times_ln_y) - sumt * sumlny) / (n * np.sum(t_squared) - sumt ** 2)
A = (mean_ln_y * np.sum(t_squared) - mean_t * np.sum(t_times_ln_y)) / (np.sum(t_squared) - n * (mean_t ** 2))
a = np.exp(A)

# Menghitung s_i (error kuadrat)
for i in range(n):
    si[i] = (ln_y[i] - A + b * t[i]) ** 2

# Membuat matriks C dan D untuk menghitung koefisien regresi
C = np.array([[n, sumt], [sumt, np.sum(t_squared)]])
D = np.array([[sumlny], [np.sum(t_times_ln_y)]])

# Menghitung nilai prediksi dari koefisien regresi sebelumnya
ybaru = np.zeros(n, float)
for i in range(n):
    ybaru[i] = a * np.exp(b * t[i])

# Menghitung Nilai S
S = np.sum(si)

# Menghitung Jumlah t dan tY
jumt = np.sum(t)
jumty = np.sum(t_times_ln_y)

# Mencetak persamaan regresi
print('ln(y(t)) =', b, 't +', A)

# Membuat DataFrame
data = {'No.': range(1, len(t) + 1), 't': t, 'y': y, 'ln(y)': ln_y, 't*ln(y)': t_times_ln_y, 't^2': t_squared, 'ln(y)^2': ln_y_squared, 's_i': si}
df = pd.DataFrame(data)

# Tampilkan seluruh nilai hasil perhitungan
print(df)
print('Nilai a =', a)
print('Nilai b =', b)
print('Nilai S =', S)
print('Jumlah t =', jumt)
print('Jumlah tY =', jumty)
```
